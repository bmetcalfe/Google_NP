```python
from neuralprophet import NeuralProphet
import pandas as pd

data_location = "https://raw.githubusercontent.com/bmetcalfe/Google_NP/main/"

df = pd.read_csv(data_location + 'T1_Data_All.csv')
```

```python
m = NeuralProphet()
metrics = m.fit(df)
```

```python
future = m.make_future_dataframe(df=df, periods=180)
forecast = m.predict(df=future)
```

```python
fig_forecast = m.plot(forecast)
```
